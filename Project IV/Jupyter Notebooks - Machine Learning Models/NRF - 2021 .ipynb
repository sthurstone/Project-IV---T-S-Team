```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
NRF=pd.read_csv("NRF.csv")
NRF.head(5)
```

```python
NRF.info()
```

```python
NRF2=NRF.rename(columns={"2020 worldwide retail sales (billions)": "2020 Sales (Billions)"})
NRF2
```

```python
NRF3=NRF2.nlargest(n=10, columns=['2020 Sales (Billions)'])
NRF3
```

```python
NRF3.plot(x ='Company', y='2020 Sales (Billions)', kind = 'bar', color='maroon', figsize=(10, 10))
plt.ylabel("2020 Sales (Billions)")
plt.xlabel("Retail Giant")
plt.title("National Retail Federation - Top Ten Retail Giants, 2020")
plt.savefig('NRF.png')
```
