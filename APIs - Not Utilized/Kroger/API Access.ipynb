REQUESTING TOKEN

```python
 def encodeData():
    message = f"{clientId}:{clientSecret}"

    message_encoded = message.encode('ascii')
    base64_bytes = base64.b64encode(message_encoded)
    base64_message = base64_bytes.decode('ascii')
    return base64_message
```

```python
def getToken():
    URL = 'https://api-ce.kroger.com/v1/connect/oauth2/token'
    authorization = f"Basic {encodeData()}"
    HEADER = {
        "User-Agent": "",
        "Authorization": authorization,
        "Content-Type": "application/x-www-form-urlencoded"
    }

    payload = {
        "grant_type": "client_credentials",
        "scope": "product.compact",
    }
    r = requests.request("POST", URL, data=payload, headers=HEADER).json()
    return r['access_token']
     
```

```python
URL = 'https://api-ce.kroger.com/v1/products?'
HEADER = {
    "User-Agent": "",
    "Authorization": f"Bearer {getToken()}",
    "Accept": "application/json",
}
payload = {
    "grant_type": "client_credentials",
    "scope": "product.compact",
}
r = requests.get(URL + "filter.term=milk", headers=HEADER).json()
```
